```python
import numpy as np
import pandas as pd
import tensorflow.compat.v1 as tf
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().any()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
classifier = tf.keras.models.Sequential()
```

```python

classifier.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
classifier.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
classifier.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam',loss = 'binary_crossentropy',metrics=['accuracy'])
```

```python
history = classifier.fit(x_train,y_train,batch_size=32,epochs=100)
```

```python
history.history??
```

```python
plt.plot(history.history["accuracy"])
plt.title("model accuracy")
plt.ylabel("accuracy")
plt.xlabel("epochs")
plt.show()
```

```python
plt.plot(history.history["loss"])
plt.title("model loss")
plt.ylabel("loss")
plt.xlabel("epochs")
plt.show()
```

```python
y_preds = classifier.predict(x_test)
y_preds = (y_preds>0.5)
np.concatenate((y_preds.reshape(len(y_preds),1),y_test.reshape(len(y_test),1)),1)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test,y_preds)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = [0, 1])
cm_display.plot()
ANN_accuracy = accuracy_score(y_test,y_preds)
print(ANN_accuracy)
```

```python
from sklearn.metrics import mean_absolute_error as mae 
ANN_mae = mae(y_test, y_preds) 
print(ANN_mae)
```

```python
from sklearn.metrics import mean_squared_error
ANN_rmse = np.sqrt(mean_squared_error(y_test, y_preds))
print(ANN_rmse)
```

```python
#Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression(random_state=0)
lg.fit(x_train,y_train)
```

```python
y_pred = lg.predict(x_test)
```

```python
np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test,y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = [0, 1])
cm_display.plot()
LR_accuracy = accuracy_score(y_test,y_pred)
print(LR_accuracy)
```

```python
from sklearn.metrics import mean_absolute_error as mae 
LR_mae = mae(y_test, y_pred) 
print(LR_mae)
```

```python
from sklearn.metrics import mean_squared_error
LR_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(LR_rmse)
```

```python
model_names = ['Artificial Neural Network.', 'Logistic Regression']
mae_values = [ANN_mae, LR_mae]

# Plot RMSE values
plt.bar(model_names, mae_values)
plt.xlabel('Classification Model')
plt.ylabel('MEA')
plt.title('Mean Absolute Error for Classification Models')
plt.show()

# Find the index of the model with the lowest RMSE
best_model_index = np.argmin(mae_values)

# Print the best model
best_model = model_names[best_model_index]
print("Best Model:", best_model)
```

```python
model_names = ['Artificial Neural Network', 'Logistic Regression']
rmse_values = [ANN_rmse, LR_rmse]

# Plot RMSE values
plt.bar(model_names, rmse_values)
plt.xlabel('Classification Model')
plt.ylabel('RMSE')
plt.title('Root Mean Square Error for Regression Models')
plt.show()

# Find the index of the model with the lowest RMSE
best_model_index = np.argmin(rmse_values)

# Print the best model
best_model = model_names[best_model_index]
print("Best Model:", best_model)
```

```python
model_names = ['Artificial Neural Network', 'Logistic Regression']
accuracy = [ANN_accuracy, LR_accuracy]

# Plot RMSE values
plt.bar(model_names, accuracy)
plt.xlabel('Classification Model')
plt.ylabel('Accuracy')
plt.title('Accuracy of model')
plt.show()

# Find the index of the model with the high accuracy
best_model_index = np.argmax(accuracy)

# Print the best model
best_model = model_names[best_model_index]
print("Best Model:", best_model)
```
